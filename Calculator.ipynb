#  CLI Calculator

This is a simple calculator built in Python .
It supports addition, subtraction, multiplication, and division.  
Built as a mini-project to practice Python basics.


```python
#define arithmetic fn
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error: cannot divide by zero!"
    return x / y

#start program
print("Welcome to the CLI calculator!")

#Infinite loop: Keeps calculator running until user decides to exit
while True:
    print("\nSelect operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")

    #Take user choice as string input
    choice = input("Enter choice (1/2/3/4/5): ")
                   
    #Exit condition if user chooses 5
    if choice == '5':
        print("Thank you for using the calculator. Goodbye! 👋")
        break #stop loop
        
    #ask for two numbers (float to allow decimal calculations)
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))

    #perform operation based on user choice 
    if choice == '1':
        print(f"Result: {a} + {b} = {add(a, b)}")
    elif choice == '2':
        print(f"Result: {a} - {b} = {subtract(a, b)}")
    elif choice == '3':
        print(f"Result: {a} * {b} = {multiply(a, b)}")
    elif choice == '4':
        print(f"Result: {a} / {b} = {divide(a, b)}")
    else:
        print("Invalid choice. Please select a valid operation (1/2/3/4/5).") #for wrong input
```
